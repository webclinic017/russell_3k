```python


def build_days(key1):
    start_date = datetime(2021, 7, 14)
    temp = list(rrule(freq=WEEKLY, byweekday=key1, count=35, dtstart=start_date))
    days_list = []
    for i in temp:
        days_list.append(str(i)[:10])
    return days_list


report_df = pd.DataFrame()

for day in days:
    month = str(day)[:7]
    path9 = Path(f"/home/gdp/russell_3k/reports/port_results/{month}/{day}/one_{label_lst[0]}.csv")
    df0 = pd.DataFrame(pd.read_csv(path9)[['date','portfolio']]).round(2)
    df0.columns = ['exit_date','portfolio'] 

    investment = df0['portfolio'].iloc[0]
    max1 = df0['portfolio'].iloc[0]
    up10_goal = max1 * 1.1
    down10_sell = max1 * 0.895
    new_max = []
    for i in df0['portfolio']:
        if i > max1:
            max1 = i
            new_max.append(i)
        else:
            new_max.append(max1)

    new_loss = []
    for i in df0['portfolio']:
        if i < max1:
            max1 = i
            new_loss.append(i)
        else:
            new_loss.append(max1)            

    df0['max'] = new_max
    df0['min'] = new_loss
    df0['max_ret'] = df0['max'] / investment
    df0['max_los'] = df0['min'] / investment
    df0['up10'] = df0['portfolio'] > up10_goal
    df0['down10'] = df0['portfolio'] < down10_sell
    df0['returns'] = ((df0['portfolio'] - investment) / investment) * 100
    df0['down_from_max'] = df0['portfolio'] / df0['max']
    df0['drop'] = df0['down_from_max'] < 0.985
    df0['over20'] = df0['returns'] >= 25.0
    up20 = df0[df0['over20'] == True]

    up10 = df0.query(f"up10 == {True}")
    up10 = up10[up10['drop'] == True]
    down10 = df0.query(f"down10 == {True}")

    if not up20.empty:
        fin = pd.DataFrame(up20.head(1).copy())
        res1 = 'W'
    
    elif not up10.empty and not down10.empty:    
        try:
            if up10.iloc[0]['exit_date'] < down10.iloc[0]['exit_date']:
                fin = pd.DataFrame(up10.head(1).copy())
                res1 = 'W'
            else:
                fin = pd.DataFrame(down10.head(1).copy())
                res1 = 'L'
        except Exception:
            print(day, '1')                
        
    elif not up10.empty:        
        try:
            fin = pd.DataFrame(up10.head(1).copy())
            res1 = 'W'
        except Exception:
            print(day, '2')

    elif not down10.empty:
        try:
            fin = pd.DataFrame(down10.head(1).copy())
            res1 = 'L'
        except Exception:
            print(day, '3')
                    
    else:
        fin = pd.DataFrame(df0.tail(1).copy())
        res1 = 'N'
            
    fin['result'] = res1
    fin.insert(loc=0, column='investment_date', value=day)
    fin.insert(loc=2, column='investment', value=investment)
    report_df = report_df.append(fin)
                
df = pd.DataFrame(report_df.copy())

mondays=build_days(MO)
tuesdays=build_days(TU)
wednesdays=build_days(WE)
thursdays=build_days(TH)
fridays=build_days(FR)

matrix = []
for i in df['investment_date']:
    if i in mondays:
        matrix.append('M')
    elif i in tuesdays:
        matrix.append('T')
    elif i in wednesdays:
        matrix.append('W')                
    elif i in thursdays:
        matrix.append('Th')
    elif i in fridays:
        matrix.append('F')
    else:
        matrix.append('-')
idx = 1
new_col = matrix
df.insert(loc=idx, column='day', value=new_col)
df = df.set_index('day').round(2)  

del df['down_from_max']
del df['drop']
del df['up10']
del df['down10']

df['gl'] = df['portfolio'] - df['investment']

df.to_csv(f"reports/score_sheet/printin_press_{label_lst[0]}.csv")
temp = pd.DataFrame(df.copy()).reset_index()
# print(df['gl'].sum())


investment = round(df['investment'].sum(),2)
exit_total = round(df['gl'].sum(),2) + investment

print("\n", df['result'].describe(), "\n")
print(f"total_investment = {'${:,.2f}'.format(investment)}")
print(f"total_exit_position = {'${:,.2f}'.format(exit_total)}")
print(f"total_return = {'{:,.2f}%'.format(((exit_total-investment)/investment)*100)}", "\n")

results_lst = list(df['result'])
total = len(results_lst)
wins = results_lst.count('W')
losses = results_lst.count('L')
neutral = results_lst.count('N')
print(
    f"Total = {total} \n\
    > Wins = {wins}, [{round(wins/total,2)}] \n\
    > Losses = {losses}, [{round(losses/total,2)}] \n\
    > Neutral = {neutral}, [{round(neutral/total,2)}]"
)

fd = df[df['result'] != 'N']
results_lst = list(fd['result'])
total = len(results_lst)
wins = results_lst.count('W')
losses = results_lst.count('L')
neutral = results_lst.count('N')
print(f"\nTotal = {total} \n\
    > Wins = {wins}, [{round(wins/total,2)}] \n\
    > Losses = {losses}, [{round(losses/total,2)}] \n\
    > Neutral = {neutral}, [{round(neutral/total,2)}]"
)
```
